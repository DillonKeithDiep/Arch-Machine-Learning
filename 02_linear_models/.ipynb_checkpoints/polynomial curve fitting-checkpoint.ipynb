### Terminology
- Training Set $\mathbf{X} = \{\mathbf{x}_1,...,\mathbf{x}_N\}$ : data used to tune the parameters of an adaptive model
- Target Vector $\mathbf{t}$ : expresses the category of an observable data entity
- Training Phase : determining the output function $\mathbf{y}(\mathbf{x})$ of the machine learning algorithm (also known as *learning phase*)
- Test Set : data to be operated upon by the trained model
- Generalisation : ability of adaptive model extrapolating for newly observed data
- Preprocessing: transforming raw data into a space of more useful new variables to improve the training process
- Feature Extraction : preprocessing to identify useful features of the observed data structure
- Exploration and Exploitation Trade-off: Choosing to look for new effective actions or choosing to optimise greedily


### Categories of Machine Learning Algorithms
- Supervised Learning: training data comprises examples of the input vectors along with the corresponding target vectors (expected output of training data is known)
- Unsupervised Learning: training data does not consist of corresponding target values (expected output is not known)
- Reinforcement Learning: An agent interacts with the environment to discover what maximises reward

### Categories of Machine Learning Problems
- Classification: assigning each input vector to one of a finite discrete categories
- Regression: predicting output of one or more continuous variables
- Clustering: group similar observed data
- Density Estimation: determining the distribution of data within the input space
- Dimensionality Reduction: projecting data from a high-dimensional space to a low-dimensional one
- Visualisation: performing dimensionality reduction for the purposes of viewing

Such categories are not mutually exclusive


### Polynomial Curve Fitting
- Suppose observation $x \in \Re$.
- Predict the value of target $t \in \Re$.
- Let data be modelled by function $sin(2 \pi x)$ with random noise.
- Given training set $\mathbf{x} \equiv (x_1,...,x_N)^T$, predict target set $\mathbf{t} \equiv (t_1,...,t_N)^T$
- Our goal is to exploit the training set to predict the value $\hat{t}$ of the target variable for some new value $\hat{x}$ of the input variable


```python
%matplotlib inline
from math import pi, sin, cos, tan, sqrt
import matplotlib.pyplot as plt

def frange(start, stop, step):
    i = start
    while i < stop:
        yield i
        i += step

f = lambda x: sin(2*pi*x)
xData = [x for x in frange(0.0, 1.0, 0.01)]
yData = [f(x) for x in xData]
xObs = [x for x in frange(0.0, 1.0, 0.1)]
yObs = [0.0, 0.506, 0.999, 0.927, 0.409, 0.132, -0.657, -0.869, -0.957, -0.488, -0.051]
plt.figure()
fn = plt.plot(xData, yData, label = 'f(x)')[0] # line2D, use first item
observation = plt.scatter(xObs, yObs, label = 'observation')
plt.legend(handles=[fn, observation])
plt.title('Target function T and observed data X')
```

A simple approach for curve fitting is to fit the data using a polynomial function of the form:
$$y(x,\mathbf{w}) = w_0 + w_1 x + w_2 x + ... + w_M x^M = \sum^M_{j=0}w_j x^j$$
- M is the *order* of the polynomial, and $x_j$ denotes the $x$ raised to the power of $j$.
- $w_0,...,w_M$ is denoted by vector $\mathbf{w}$.
- Polynomial function $y(x,\mathbf{w})$ is a nonlinear function of $x$, but a linear function of the coefficients $\mathbf{w}$.
- The values of the coefficients can be determined by fitting the polynomial to the training data through minimising the *error function*
- The error function measures the misfit between the function $y(x, \mathbf{w})$ for any given value of $\mathbf{w}$ with the training set data points. 
- A common error function is the sum of the squares of the errors between the predictions $y(x_n, \mathbf{w})$ for each data point $x_n$ and the corresponding target values $t_n$:
$$E(\mathbf{w} = \frac{1}{2}\sum{N}{n=1}\{ y(x_n, \mathbf{w}) - t_n \}^2$$
- The curve fitting problem is solved by choosing the value $\mathbf{w}$ where $E(\mathbf{w})$ is as minimised.
- As the error function is a quadratic function of the coefficients $\mathbf{w}$, its derivatives with respect to the coefficients will be linear in the elements of $\mathbf{w}$, and so the minimisation of the error function has a unique solution, denoted $\mathbf{w}^*$.
- The resulting polynomial is given by the function $y(x, \mathbf{w}^*)$.

Let M be the order of the polynomial. When $(M = 0)$ and $(M = 1)$, the model fits poorly to the data, as they are not representative of the true function $sin(\pi x)$. A high order polynomial such as $(M=9)$ fits the data very well but oscillates significantly - overfitting the data precisely with noise. This in fact detracts from representing the true function in comparison to a more balanced order such as $(M=3)$.

For each choice of M, we can evaluate the residual value $E(\mathbf{w}^*)$ given by the sum of squares error function for the training data.$E(\mathbf{w}^*)$ can also be evaluated for the test data set.

The root-mean-square (RMS) error is defined
$$E_{RMS}=\sqrt{2E(\mathbf{w}^*)/N}$$
the division of $N$ allows comparing different sizes of data sets on equal footing. The square root ensures measurement on the same scale (and units) as the target variable t.

The test set error is a measure of how well a model predicts the values of $t$ for new data observations of $x$.

```python
import numpy as np
m0 = np.polyfit(xObs, yObs, 0)
m1 = np.polyfit(xObs, yObs, 1)
m3 = np.polyfit(xObs, yObs, 3)
m10 = np.polyfit(xObs, yObs, 10)

m0e = np.poly1d(m0)
m1e = np.poly1d(m1)
m3e = np.poly1d(m3)
m10e = np.poly1d(m10)

plt.figure()
plt.ylim(-1,1)
fn = plt.plot(xData, yData, label = 'fn(x)')[0]
observation = plt.scatter(xObs, yObs, label = 'observation')
hm0 = plt.plot(xData, m0e(xData), 'r', label = 'M=0')[0]
hm1 = plt.plot(xData, m1e(xData), 'g', label = 'M=1')[0]
hm3 = plt.plot(xData, m3e(xData), 'b', label = 'M=3')[0]
hm10 = plt.plot(xData, m10e(xData), 'y', label = 'M10')[0]
plt.legend(handles=[fn, observation, hm0, hm1, hm3, hm10])
plt.title('Prediction of T with M coefficients')
```

```python
from sklearn.metrics import mean_squared_error

rmsObs0 = sqrt(mean_squared_error(yObs, m0e(xObs)))
rmsObs1 = sqrt(mean_squared_error(yObs, m1e(xObs)))
rmsObs3 = sqrt(mean_squared_error(yObs, m3e(xObs)))
rmsObs10 = sqrt(mean_squared_error(yObs, m10e(xObs)))
rmsData0 = sqrt(mean_squared_error(yData, m0e(xData)))
rmsData1 = sqrt(mean_squared_error(yData, m1e(xData)))
rmsData3 = sqrt(mean_squared_error(yData, m3e(xData)))
rmsData10 = sqrt(mean_squared_error(yData, m10e(xData)))
plt.figure()
test = plt.plot([0, 1, 3, 10],[rmsData0, rmsData1, rmsData3, rmsData10],
    'r', label = 'Test')[0]
training = plt.plot([0, 1, 3, 10], [rmsObs0, rmsObs1, rmsObs3, rmsObs10],
    'b', label = 'Training')[0]
plt.legend(handles=[training, test])
plt.title('RMSE')
```

- The X axis represents the number of coefficients for the prediction (M)
- As M increases, the RMSE for the training data tends towards 0
- Overfitting causes the RMSE for test data (value difference from true function at unobserved points) to soar for higher M values
- A higher order polynomial contains all lower order polynomials, but as M increases, the magnitude of coefficients typically gets larger, leading to significant fluctuation
- Dense data collection reduces overfitting, but could be inefficient compared to a smooth solution

- The number of parameters is not necessarily the most appropriate measure of model complexity

```python
import random;
xObs2 = [x for x in frange(0.0, 1.0, 0.05)]
yObs2 = [f(x)+random.uniform(0, 0.1) for x in frange(0.0, 1.0, 0.05)]
m10_2 = np.polyfit(xObs2, yObs2, 10)
m10e_2 = np.poly1d(m10_2)

plt.figure()
plt.ylim(-1,1)
fn = plt.plot(xData, yData, label = 'fn(x)')[0]
observation = plt.scatter(xObs, yObs, label = 'observation (sparse)')
observation2 = plt.scatter(xObs2, yObs2, label = 'observation (dense)')
hm10 = plt.plot(xData, m10e(xData), 'y', label = 'M10 (sparse)')[0]
hm10_2 = plt.plot(xData, m10e_2(xData), 'y', label = 'M10 (dense)')[0]
plt.legend(handles=[fn, observation, observation2, hm10, hm10_2])
plt.title('Sparse vs Dense observed dataset for M = 10')
```

- the least squares approach to finding the model parameters represents a specific case of *maximum likelihood*
- Adopting a *Bayesian* approach can avoid the over-fitting problem
    - In a Bayesian model, the *effective* number of parameters adapts automatically to the size of the data set
- *Regularisation* is a technique used to control the over-fitting phenomenon by adding a penalty term to the error function
$$\tilde{E}(\mathbf{w}) = \frac{1}{2}\sum{N}{n=1}
\{ y(x_n, \mathbf{w}) - t_n \}^2 +
\frac{\lambda}{2}||\mathbf{w}||^2,\\
||\mathbf{w}||^2 = \mathbf{w}^T\mathbf{w} = w^2_0+w^2_1+w^2_0...+w^2_M$$
- The coefficient $\lambda$ governs the relative importance of the regularisation term compared with the sum-of-squares error term
- Often the coefficient $w_0$ is omitted from the regulariser because its inclusion causes the results to depend on the choice of origin for the target variable (Hastie et al., 2001), or regularised itself
- The error function can be minimised exactly in closed form
- Techniques like this in statistics is known as *shrinkage*
    - The particular case of a quadratic regulariser is called *ridge regression* (Hoerl and Kennard, 1970).
    - In a neural network, this approach is known as *weight decay*
- The higher the regularising coefficient $\lambda$ is, the more suppression is imposed on overfitting
    - Too large or too small would lead to poor fit again
    - Again, the performance can be measured by evaluating the error
